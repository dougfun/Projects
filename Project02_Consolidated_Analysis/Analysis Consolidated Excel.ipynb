### Gathering data

#### By segment

```python
columns = ['valor_total_desconto', 'custo_total', 'lucro']
df_segmento = df.groupby('segmento')[columns].sum().round(2) # Could also use sum(), min(), etc
```

```python
type(df_segmento)
```

```python
df_segmento['margem_lucro'] = round((df_segmento['lucro'] / df_segmento['valor_total_desconto']) * 100, 2)
```

```python
df_segmento
```

#### By product

```python
columns = ['valor_total_desconto', 'custo_total', 'lucro']
df_produto = df.groupby('produto')[columns].sum().round(2)
df_produto['margem_lucro'] = round((df_produto['lucro'] / df_produto['valor_total_desconto']) * 100, 2)
df_produto
```

#### By country

```python
columns = ['valor_total_desconto', 'custo_total', 'lucro']
df_pais = df.groupby('pais')[columns].sum().round(2)
df_pais['margem_lucro'] = round((df_pais['lucro'] / df_pais['valor_total_desconto']) * 100, 2)
df_pais
```
